```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
%matplotlib inline

x = np.linspace(0, 10, 150)
y = np.sin(3*x)
f = np.exp(-x**2)
```

```python
plt.xlabel(sp.latex("Variable independiente"))
plt.ylabel(sp.latex("Velocidad del auto"))
plt.plot(x, y, linewidth=3, color="red", label="Funcin")
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(x, y, linewidth=4, linestyle=":", color=(1, 0.5, 0.5))
plt.plot(x, f, linewidth=2, linestyle="dashdot", color="dodgerblue")
```
