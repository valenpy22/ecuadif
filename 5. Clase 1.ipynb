```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X, Y = [2, 3]
U, V = [1, 5]

#Indica la dirección
plt.quiver(X, Y, U, V)
plt.grid()
```

```python
X, Y = [[2, 2], [2, 3]]
U, V = [[1, 2], [1, 1]]
plt.grid()
plt.quiver(X, Y, U, V)
```

```python
X = np.linspace(-1, 1, 3)
Y = np.linspace(-1, 1, 3)

X, Y = np.meshgrid(X, Y)
X, Y
```

```python
X, Y = np.meshgrid(np.linspace(-4, 4, 9), np.linspace(-4, 4, 9))

U = 1
V = 2*Y

#Se normalizan los vectores
N = np.sqrt(U**2 + V**2)

U = U/N
V = V/N

plt.quiver(X, Y, U, V)
plt.xlim([-5,5])
plt.ylim([-5,5])

plt.xlabel("x")
plt.ylabel("y")
plt.grid()
```

```python
x, y = np.meshgrid(np.linspace(-5, 5, 12), np.linspace(-5, 5, 12))

u = 1

#Se despeja dx/dy
v = x - y

#Norma del vector
n = np.sqrt(u**2 + v**2)

#Se normaliza el vector
u = u/n
v = v/n

plt.quiver(x, y, u, v)
plt.xlim([-5, 5])
plt.ylim([-5, 5])

plt.xlabel("x")
plt.ylabel("y")
plt.grid()

a = np.linspace(-5, 5, 200)

f = np.exp(-a) + a - 1
plt.plot(a, f, label="f")

g = 2*np.exp(-a) - a + 1
plt.plot(a, g, label="g")
plt.legend()
```

```python
x, y = np.meshgrid(np.linspace(-5, 5, 20), np.linspace(-5, 5, 20))

u = 1
v = (y-3)*np.log(y**2+1)

n = np.sqrt(u**2 + v**2)
u = u/n
v = v/n

plt.quiver(x, y, u, v)
plt.xlim([-5, 5])
plt.ylim([-5, 5])
plt.grid()
```

```python
x, y = np.meshgrid(np.linspace(0, 10, 20), np.linspace(0, 1, 20))

u = 1
v = (y**2 + 3*y - 4)*np.sqrt(y)

n = np.sqrt(u**2 + v**2)
u = u/n
v = v/n

plt.quiver(x, y, u, v)
plt.xlim([0, 10])
plt.ylim([0, 1])
plt.grid()

#y = 0, nodo
#y = 1, repulsor
```
