```python
#Se genera una partición finita de números.
#(número inicial, número final, cada cuánto genera un punto)
x = np.arange(-3.14, 3.14, 0.1)

y = np.sin(x)
plt.plot(x, y)
```

```python
#Genera 100 elementos entre 0 y 2.
x = np.linspace(0, 20, 100)
y = np.sin(x)
#y = np.sin(np.sqrt(x))


plt.plot(x, y)
```

```python
x = np.linspace(-2*pi, 2*pi, 100)
y = np.cos(x)
f = np.exp(-x**2)

plt.plot(x, y)
plt.plot(x, f)
```
